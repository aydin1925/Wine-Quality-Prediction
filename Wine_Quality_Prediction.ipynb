```python
## Matplotlib ve Seaborn import etme ve kullanma

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
## bağımlı ve bağımsız değişkenleri ayırma

x = df.drop("quality", axis = 1)
y = df["quality"]
```

```python
## veriyi %80 eğitim ve %20 test olarak bölme

x_train , x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
## Boyutlarına bakma

x_train.shape , x_test.shape , y_train.shape , y_test.shape
```

```python
## Rastgele Orman Yöntemini import etme ve kullanma

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score , classification_report
```

```python
## Rastgele Orman modelini kullanma

rastgele_orman = RandomForestClassifier(random_state=42)
rastgele_orman.fit(x_train, y_train)
```

```python
## tahminde bulunma

y_tahmin = rastgele_orman.predict(x_test)
y_tahmin ## tahmin edilen değerleri görelim
```

```python
## performansı değerlendirme

dogruluk_orani = accuracy_score(y_test, y_tahmin)
sinif_raporu = classification_report(y_test, y_tahmin)
```

```python
## ekrana yazma

print("Modelin doğuruluk oranı: " , dogruluk_orani) ## Modelin doğuruluk oranı:  0.659375
print("Sınıflandırma Raporu: \n", sinif_raporu)
```

```python
## Grid Search ile Hiperparametre Optimizasyonu yapma

from sklearn.model_selection import GridSearchCV
```

```python
## parametreleri belirleyelim

params = {'max_depth': [10,20,30], 'min_samples_split': [2,5,10]}
```

```python
gridArama = GridSearchCV(estimator=rastgele_orman, param_grid=params, cv=5).fit(x_train, y_train)
```

```python
## Sonucları ekrana yazalım

print("En iyi parametreler: ", gridArama.best_params_)
print("En yüksek doğruluk: ", gridArama.best_score_) ## En yüksek doğruluk:  0.6849601715686274
```
